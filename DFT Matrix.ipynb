Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

Define Variables

```python
#size of DFT
n = 128
#fundamental frequency
w = np.exp(-1j * 2 * np.pi / n)
```

Compute DFT matrix and plot imaginary part

```python
J,K = np.meshgrid(np.arange(n),np.arange(n))
DFT = np.power(w,J*K)
DFT = np.real(DFT)
       
plt.imshow(DFT)
plt.show()
```
